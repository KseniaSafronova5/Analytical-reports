```python
import pandas as pd
import numpy as np
import os

# Настройки
path = r'C:\Users\kseni\Desktop\Мои услуги\Автоматический отчёт'
file_name = 'sales_january_2026.csv'
full_path = os.path.join(path, file_name)

# Создаем папку, если она еще не существует
if not os.path.exists(path):
    os.makedirs(path)

# Генерация данных
np.random.seed(42)
dates = pd.date_range(start='2026-01-01', end='2026-01-31', periods=300)
shops = ['Север', 'Юг', 'Центр']
products = ['Кофемашина', 'Ноутбук', 'Смартфон', 'Наушники', 'Монитор']

data = []

for date in dates:
    shop = np.random.choice(shops)
    product = np.random.choice(products)
    
    # Логика аномалии: по воскресеньям (dayofweek == 6) продажи в 5-10 раз ниже
    if date.dayofweek == 6:
        amount = np.random.uniform(500, 2000)
    else:
        amount = np.random.uniform(5000, 50000)
        
    margin = amount * np.random.uniform(0.1, 0.3) # Маржа 10-30%
    
    data.append([date.date(), shop, product, round(amount, 2), round(margin, 2)])

# Сборка DataFrame
df = pd.DataFrame(data, columns=['Дата', 'Магазин', 'Товар', 'Сумма_чека', 'Маржа'])

# Сохранение
df.to_csv(full_path, index=False, encoding='utf-8-sig')

print(f"Файл готов и лежит тут: {full_path}")
```
