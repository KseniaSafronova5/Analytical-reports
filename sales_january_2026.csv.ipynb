```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

def generate_synthetic_sales():
    # Настройки
    start_date = datetime(2026, 1, 1)
    end_date = datetime(2026, 1, 31)
    
    # Список товаров по категориям
    categories = {
        'Электроника': ['Смартфон', 'Ноутбук', 'Наушники', 'Зарядка'],
        'Аксессуары': ['Чехол', 'Стекло', 'Ремешок'],
        'Услуги': ['Настройка', 'Страховка', 'Установка ПО']
    }
    
    data = []
    current_date = start_date
    
    while current_date <= end_date:
        # В воскресенье (weekday == 6) делаем мало продаж для аномалии
        is_sunday = current_date.weekday() == 6
        num_sales = random.randint(2, 5) if is_sunday else random.randint(15, 30)
        
        for _ in range(num_sales):
            cat = random.choice(list(categories.keys()))
            item = random.choice(categories[cat])
            
            # Сумма чека: в воскресенье чеки тоже сделаем поменьше
            if is_sunday:
                revenue = random.randint(500, 2000)
                margin = revenue * random.uniform(0.1, 0.2)
            else:
                revenue = random.randint(2000, 50000)
                margin = revenue * random.uniform(0.15, 0.35)
                
            data.append({
                'Дата': current_date.strftime('%Y-%m-%d'),
                'Товар': item,
                'Категория': cat,
                'Сумма_чека': int(revenue),
                'Маржа': int(margin)
            })
            
        current_date += timedelta(days=1)
    
    # Создаем DataFrame
    df = pd.DataFrame(data)
    
    # Сохраняем в CSV
    file_name = "sales_january_2026.csv"
    df.to_csv(file_name, index=False, encoding='utf-8-sig')
    print(f"✅ Файл {file_name} успешно создан! В нем {len(df)} строк.")

if __name__ == "__main__":
    generate_synthetic_sales()
```
